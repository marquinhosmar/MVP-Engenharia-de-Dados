**O investimento p√∫blico segue um padr√£o uniforme ao longo do tempo ou apresenta varia√ß√µes?**

An√°lise agregada (Brasil)

```python
invest_tempo = (
    df
    .groupBy("Ano")
    .agg(
        F.sum("Investimento_Real").alias("Investimento_Real_Total")
    )
    .orderBy("Ano")
)

display(invest_tempo)
```

Outro gr√°fico:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar a GOLD
df_gold = spark.table("workspace.mvp.gold_dca")

# Converter para pandas (dados agregados, volume pequeno)
pdf = df_gold.select(
    "uf",
    "Ano",
    "Receita_Total_Real",
    "Despesa_Total_Real",
    "Investimento_Real"
).toPandas()
```

Gr√°fico 1 ‚Äî Receita Total √ó Investimento (REAL)

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=pdf,
    x="Receita_Total_Real",
    y="Investimento_Real",
    hue="uf",
    alpha=0.7
)

sns.regplot(
    data=pdf,
    x="Receita_Total_Real",
    y="Investimento_Real",
    scatter=False,
    color="red"
)

plt.title("Rela√ß√£o entre Receita Total Real e Investimento Real")
plt.xlabel("Receita Total Real (R$)")
plt.ylabel("Investimento Real (R$)")
plt.grid(True)
plt.show()
```

‚úÖ Parte 4 ‚Äî Pergunta 3

**Estados com maiores gastos brutos s√£o os que mais investem em termos percentuais?**

üìä Ranking percentual m√©dio (real)

```python
ranking_pct = (
    df
    .groupBy("uf")
    .agg(
        F.avg("Investimento_Real_pct_Despesa_Real").alias("Pct_Investimento_Medio")
    )
    .orderBy(F.desc("Pct_Investimento_Medio"))
)

display(ranking_pct)
```

üìä An√°lise do Ranking M√©dio do Percentual de Investimento

(Investimento / Despesa Total ‚Äì m√©dia do per√≠odo 2013‚Äì2024)

Diferentemente de rankings absolutos, este indicador mede o esfor√ßo relativo de investimento, isto √©, quanto do or√ßamento total cada estado direciona para investimentos.

2Ô∏è‚É£ Principais evid√™ncias observadas:

üîπ Alta heterogeneidade entre os estados

Observa-se uma varia√ß√£o expressiva no percentual m√©dio de investimento:
- Estados no topo do ranking investem, em m√©dia, entre 8% e 10% de suas despesas totais;
- Estados na parte inferior investem menos de 3%, em alguns casos pr√≥ximos de 2%.
- Isso indica que o esfor√ßo de investimento n√£o √© uniforme entre os entes federativos.

üîπ Estados que mais gastam n√£o s√£o necessariamente os que mais investem proporcionalmente

- Comparando esse ranking com os gr√°ficos de despesa total absoluta, observa-se que:

Conclus√£o importante:
- Maior gasto absoluto n√£o implica maior prioridade relativa ao investimento.

üîπ Evid√™ncia de diferentes estrat√©gias or√ßament√°rias

O ranking sugere a exist√™ncia de estrat√©gias distintas de aloca√ß√£o or√ßament√°ria:

‚úÖ Parte 5 ‚Äî Pergunta 4

**Estados priorizam despesas correntes em detrimento dos investimentos ao longo do tempo?**

üìä Evolu√ß√£o do peso das despesas correntes


```python
corrente_tempo = (
    df
    .groupBy("Ano")
    .agg(
        F.avg("Despesa_Corrente_Real_pct_Despesa_Real")
        .alias("Pct_Despesa_Corrente")
    )
    .orderBy("Ano")
)

display(corrente_tempo)
```

üìà An√°lise da Evolu√ß√£o do Percentual de Despesas Correntes (2013‚Äì2024)

Esse indicador mede o grau de rigidez or√ßament√°ria, uma vez que despesas correntes (pessoal, custeio, previd√™ncia) tendem a ser de dif√≠cil compress√£o no curto e m√©dio prazo.
